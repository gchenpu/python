### Reshaping arrays

````python
array1 = np.arange(10)
print(array1.shape)
print(array1)

array1 = np.reshape(array1, (5,2))
print(array1.shape)
print(array1)
````

### Matrix multiplication
Elementwise mutiplication: A*B  
Matrix product: numpy.dot(A,B), or A @ B

````python
A = np.array(( [1,2], [3,4] ), dtype=float)
B = np.array(( [5.,6], [7,8] ))
print(A*B) # elementwise mutiplication
print(np.dot(A,B)) # matrix product
print(A @ B)

print(A.mean(), \
      A.sum(), \
      A.min(), \
      A.max())

print(np.mean(A), \
      np.sum(A), \
      np.min(A), \
      np.max(A))
````

### Indexing, slicing, iterating
To index a NumPy array, using square brackets [i,j].

````python
C = np.arange(10)

print(C, C.shape)
print(C[0:5]) # prints first five elements (remember zero-based indexing!)
print(C[::-1])
print(C[::2])
print(C[:,None], '\n', C[:,None].shape) # adding a column of size 1
````

````python
E = np.arange(50).reshape(5,10)
print(E)
print(E[0:5,6:10]) # first 5 rows, last 5 columns

#Indexing over multiple axes is done with respect to first axis [i]:
for row in E:
    print(row)
````

### Reshaping versus resizing
The **```numpy.reshape()```** function returns a *new* argument with the specified shape.
The **```numpy.resize()```** function resizes the array *in place*.

````python
F = np.linspace(0,10,50)
print(F.reshape(10,5))
print(F.shape)

F.resize(10,5)
print(F)
print(F.shape)
````

## SciPy
https://docs.scipy.org/doc/scipy-0.17.0/reference/tutorial/

### Intro to SciPy using the ```scipy.stats``` module
using numpy.random.normal(), which draws random samples around a Gaussian distribution taking 3 arguments as input (location, stdev/scale, and size): http://docs.scipy.org/doc/numpy-1.10.1/reference/generated/numpy.random.normal.html

````python
import scipy.stats

N = 1000
xvals = np.linspace(1,100,N)

a_signal = np.linspace(1,100,N)
a_noise = np.random.normal(loc=0, scale=5, size=N)
a = a_signal+a_noise

b_signal = np.linspace(1,100,N)
b_noise = np.random.normal(loc=0, scale=15, size=N)
b = b_signal+b_noise
````

````python
a_ctd = a - a.mean()
b_ctd = b - b.mean()

a_stdev = np.std(a, ddof=1) # ensures 1/(N-1), default is (1/N)
#a_stdev = ( (1./(N-1)) * numpy.sum(a_ctd**2.) ) ** 0.5
#a_stdev = numpy.sqrt( (1./(N-1)) * numpy.sum(a_ctd**2.) )

b_stdev = np.sqrt( (1./(N-1)) * np.sum(b_ctd**2.) )

print(a_stdev, b_stdev)
````

### Pearson correlation using ```scipy.stats.pearsonr()```

Compute the correlation between a and b.  You can do this using **```scipy.stats.pearsonr()```**.

Note that this function outputs a tuple with the correlation value and the p-value.  See the documentation here:  http://docs.scipy.org/doc/scipy-0.16.0/reference/generated/scipy.stats.pearsonr.html

Also beware that the **```scipy.stats.pearsonr()```** function does not use a conservative $\frac{1}{(N-1)}$ estimate of standard deviation.

$$ \mathrm{corr} = \frac{ \sum^n_{i=1} (a_i-\bar{a})(b_i-\bar{b})}{\sigma_a \sigma_b} $$

````python
ab_corr = scipy.stats.pearsonr(a,b)
print(ab_corr)

a_stdev, b_stdev = np.std(a), np.std(b) # note multiple assignments per line, NON-conservative estimate
ab_corr = np.mean( (b_ctd*a_ctd) / (a_stdev*b_stdev) )
print(ab_corr)
````

### Linear regression using ```scipy.stats.linregress()```

Now calculate a simple linear regression on the a and b arrays.  Note **```scipy.stats.linregress()```** outputs 5 different variables.  See its documentation here:  http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html

````python
a_slope, a_intercept, a_rval, a_pval, a_stderr = scipy.stats.linregress(a,xvals)
b_slope, b_intercept, b_rval, b_pval, b_stderr = scipy.stats.linregress(b,xvals)
print(a_slope, b_slope)

a_fit = a_slope * xvals + a_intercept
b_fit = b_slope * xvals + b_intercept
````

### Matlabplotlib
One way is using **```plt.plot()```** directly;  
Another way is calling **```plt.subplot()```**, which will allow you to plot panels using a (row,col,plot) syntax:

````python
plt.plot(xvals, a, label='a')
plt.plot(xvals, a_fit, label='a fit')
plt.legend()
````

````python
# subplot
ax1 = plt.subplot(1,1,1) # (rows, cols, plot)
ax1.plot(xvals, a, label='a')
ax1.plot(xvals, a_fit, label='a fit')
ax1.legend()
````

### Creating publication-ready figures
The most flexible way of creating a figure is to create it using **```fig = plt.figure()```** and *adding* subplots one by one using **```fig.add_subplot()```**.

** The advantage of this method is that axes can be adjusted individually and there is a LOT of flexibitily here.**

**If you plan to be creating publication-ready figures, this is a great place to start.**

Note this figure is saved twice as a PDF using the **```plt.savefig()```** function:  http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.savefig

````python
fig = plt.figure(figsize=(10,6)) # size is optional

ax1 = fig.add_subplot(2,1,1) # (rows, cols, plot)
ax1.plot(xvals, a, label='a', color='black', lw=2)
ax1.plot(xvals, a_fit, color='red', label='a fit', lw=3)
ax1.set_ylim(-50,150)
ax1.set_ylabel('a label')
ax1.legend(loc=0)

ax2 = fig.add_subplot(2,1,2)
ax2.plot(xvals, b, label='b', color='black', lw=2)
ax2.plot(xvals, b_fit, color='red', label='b fit', lw=3)
ax2.set_ylim(-50,150)
ax2.set_ylabel('b label')
ax2.legend(loc=0)

plt.savefig('ab_trends.pdf', transparent=True)

plt.tight_layout() # helpful for fitting plot into pdf
plt.savefig('ab_trends_tight_layout.pdf', transparent=True)
````
